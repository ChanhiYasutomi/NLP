    kana=False: カタカナを半角に変換しない（True にすると変換される）。
    digit=False: 数字を半角に変換しない（True にすると変換される）。

    このコードは、MeCabを使用して日本語のテキストを形態素解析し、解析結果をノードとして取得しています。以下にコードの各行の説明を示します。

    m = MeCab.Tagger(): MeCabの Tagger クラスのインスタンスを作成します。Tagger クラスは形態素解析を行うための主要なクラスであり、このインスタンスを使ってテキストを解析します。
    node = m.parseToNode(text): parseToNode メソッドを使用して、与えられたテキストを形態素解析の対象にします。
    parseToNode メソッドは、解析結果の最初のノードを取得します。
    このノードから順にたどりながら解析結果を取得することができます。

    簡単に言えば、これらの行はMeCabを使って与えられたテキストを形態素解析し、解析結果をノードとして取得するための初期化を行っています。
    その後のループや処理で各ノードをたどり、必要な情報を取得することが一般的です。

```python
import MeCab

text = "私は昨日、美味しいカレーライスを食べました。"

m = MeCab.Tagger()
node = m.parseToNode(text)

while node:
  print(node.surface)
  node = node.next
```

```python
f = open("/content/nlp_datascience_cram_school_1.txt")
text = f.read()
f.close()

m = MeCab.Tagger()
node = m.parseToNode(text)

meishi_list = []
doshi_list = []
keiyoshi_list = []

while node:
    if node.feature.split(",")[0] == "名詞":
        meishi_list.append(node.surface)
    elif node.feature.split(",")[0] == "動詞":
        doshi_list.append(node.feature.split(",")[6])
    elif node.feature.split(",")[0] == "形容詞":
        keiyoshi_list.append(node.feature.split(",")[6])

    node = node.next
```

```python
display(meishi_list[:10])
display(doshi_list[:10])
display(keiyoshi_list[:10])
```

```python
m = MeCab.Tagger()
node = m.parseToNode(text)

meishi_list = []
doshi_list = []
keiyoshi_list = []

while node:
    if node.feature.split(",")[0] == "名詞":
        meishi_list.append(node.surface)
    elif node.feature.split(",")[0] == "動詞":
        doshi_list.append(node.surface)
    elif node.feature.split(",")[0] == "形容詞":
        keiyoshi_list.append(node.surface)

    node = node.next
```

```python
display(meishi_list[:10])
display(doshi_list[:10])
display(keiyoshi_list[:10])
```

### 頻出単語の抽出

```python
import collections
c_meishi = collections.Counter(meishi_list)
c_doshi = collections.Counter(doshi_list)
c_keiyoshi = collections.Counter(keiyoshi_list)

display(c_meishi.most_common(10))
display(c_doshi.most_common(10))
display(c_keiyoshi.most_common(10))
```

    このコードは、Pythonの標準ライブラリである collections モジュールを使用して、名詞のリスト meishi_list に含まれる単語の出現回数を数え、
    最も出現頻度が高い上位10件を表示する処理を行っています。

    以下にコードの各部分を説明します。

    import collections: Pythonの collections モジュールをインポートします。
    このモジュールには、データのコレクションを操作するためのクラスや関数が含まれています。

    c = collections.Counter(meishi_list): Counter クラスを使用して、meishi_list の各要素（単語）の出現回数を数え、辞書のような形で保存します。
    c は Counter オブジェクトです。

    c.most_common(10): most_common メソッドを使用して、出現頻度が高い順にソートされた上位10件の要素（単語とその出現回数）を取得します。
    結果はリストとして返されます。

### ワードクラウドの作成

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
!apt-get install -y fonts-ipaexfont
```

```python
wordcloud = WordCloud(font_path="/usr/share/fonts/truetype/ipaexg.ttf",  # インストールされたフォントのパス
                      background_color="white",
                      width=600,
                      height=400,
                      min_font_size=10,
                      prefer_horizontal=1.0)

wordcloud.generate(" ".join(meishi_list))

plt.figure(figsize = (12,8))
plt.imshow(wordcloud)
plt.axis("off")
```

    このコードは、WordCloudを使用して名詞のリスト meishi_list からワードクラウド（単語の出現頻度を視覚化したもの）を生成し、Matplotlibを使用してそのワードクラウドを表示する処理を行っています。
    以下にコードの各部分を説明します。

    wordcloud = WordCloud(...): WordCloud オブジェクトを作成します。
    このオブジェクトは、ワードクラウドを生成するための様々なパラメータを指定することができます。
    この例では以下のように指定されています。

    font_path: 使用するフォントのパス。ここでは "ipaexg.ttf" を使用しています。
    background_color: ワードクラウドの背景色。ここでは白色に設定されています。
    width: ワードクラウドの幅。
    height: ワードクラウドの高さ。
    min_font_size: 表示する単語の最小フォントサイズ。
    prefer_horizontal: 水平に配置される単語の比率。1.0だとすべて水平。
    wordcloud.generate(" ".join(meishi_list)): WordCloud オブジェクトに対して、名詞のリスト meishi_list を空白で結合して文字列にし、それをもとにワードクラウドを生成します。
    generate メソッドに文字列を渡すことで、その文字列内の単語の出現頻度を元にワードクラウドが作成されます。
    plt.figure(figsize=(12, 8)): Matplotlibの figure 関数を使用して、図（Figure）のサイズを設定します。
    ここでは横幅が12、縦幅が8に設定されています。
    plt.imshow(wordcloud): Matplotlibの imshow 関数を使用して、WordCloud オブジェクトを表示します。
    plt.axis("off"): 軸を非表示にします。これにより、軸のラベルや目盛りが表示されません。

    これにより、名詞のリストから生成されたワードクラウドが Matplotlib を使用して表示されます。ワードクラウドは単語の出現頻度に応じて単語のサイズが調整され、より頻繁に出現する単語が大きく表示されます。

```python
wordcloud = WordCloud(font_path="/usr/share/fonts/truetype/ipaexg.ttf",  # インストールされたフォントのパス
                      background_color="white",
                      width=600,
                      height=400,
                      min_font_size=10,
                      prefer_horizontal=1.0)

wordcloud.generate(" ".join(doshi_list))

plt.figure(figsize = (12,8))
plt.imshow(wordcloud)
plt.axis("off")
```
